```python
# dimensional reduction (PCA)을 할 것이다. 
# SVM = classification (baseline 성능, 비교를 위해서)

# decision tree = classification
# random forest = classification

# decision tree = regression
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import tree
from sklearn import svm

from keras.datasets import mnist
# MNIST dataset
# input = 28 x 28 image => 784 vector로 변환
# training = 60,000 x 784 / test = 10,000 x 784
# 손글씨 이미지 데이터셋 784 x 1짜리 벡터로 변환 = 28 x 28 pixel로 이루어진 손글씨 이미지
((x_train, y_train), (x_test, y_test)) = mnist.load_data()

# X_train = (60000, 28, 28) 60000개의 데이터 28 x 28, 3차원 
X_train = x_train.reshape(-1,28*28) # row의 개수를 그대로, column의 개수를 28 x 28로 바꾼다 (60,000 x 784)
X_test = x_test.reshape(-1,28*28)

N_train,D = X_train.shape  # 60,000, 784
N_test,D = X_test.shape    # 10,000, 784

# dimensional reduction (PCA)
# covariance
mu = np.mean(X_train, axis = 0)
C = X_train - mu
C = (1/N_train) * C.T.dot(C)

# SVD: C = U*S*V^T
U, s, V = np.linalg.svd(C, full_matrices = True)
S = np.diag(s)
V = V.T

# 전체 입력 차원 784 중에서, 차원을 축소해서 사용
D_reduce = 30
X_train = X_train.dot(V[:,0:D_reduce]) # N_train x D_reduce
X_test = X_test.dot(V[:,0:D_reduce]) # N_train x D_reduce
```

```python
# SVM = classification
model_svm = svm.SVC(kernel = 'rbf')
model_svm.fit(X_train,y_train)

y_svm_train = model_svm.predict(X_train)
y_svm_test = model_svm.predict(X_test)

print(np.sum(y_svm_train == y_train) / N_train) # training accuracy, SVM
print(np.sum(y_svm_test == y_test) / N_test) # test accuracy, SVM
```

```python
# decision tree = classification
model_tre = tree.DecisionTreeClassifier(max_depth = 10, min_samples_split = 1000, min_samples_leaf = 1000)
model_tre.fit(X_train,y_train)

y_tre_train = model_tre.predict(X_train)
y_tre_test = model_tre.predict(X_test)

print( np.sum(y_tre_train == y_train) / N_train ) # training accuracy, decision tree
print( np.sum(y_tre_test == y_test) / N_test ) # test accuracy, decision tree
```

```python
# random forest = classification
from sklearn.ensemble import RandomForestClassifier

N_set = np.array([1, 2, 5, 10, 20, 50, 100])
accuracy_train = []
accuracy_test = []

for N in N_set:
    model_rdf = RandomForestClassifier(n_estimators = N, max_depth = 10, min_samples_split = 1000, min_samples_leaf = 1000)
    model_rdf.fit(X_train,y_train)

    y_rdf_train = model_rdf.predict(X_train)
    y_rdf_test = model_rdf.predict(X_test)

    print(np.sum(y_rdf_train == y_train) / N_train) # training accuracy, random forest
    print(np.sum(y_rdf_test == y_test) / N_test) # test accuracy, random forest
    
    accuracy_train.append(np.sum(y_rdf_train == y_train) / N_train)
    accuracy_test.append(np.sum(y_rdf_test == y_test) / N_test)
    
plt.figure()
plt.plot(N_set,accuracy_train)
plt.plot(N_set,accuracy_test)
```
