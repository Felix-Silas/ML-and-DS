```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,6,100)
np.random.shuffle(x) # 순서 섞기

y = 5 + x - 2*x**2 + 1e-1*x**3 + 1e-2*x**5
y = y + 1e1*np.random.randn(100)

x = x.reshape(-1,1)
y = y.reshape(-1,1)

plt.figure()
plt.scatter(x,y,s = 10, c = 'gray')
```

```python
# regulariztion 없이 한 경우
N,D = x.shape

sse = np.zeros(20)
sse[:] = np.nan

theta_norm2 = np.zeros(20)
theta_norm2[:] = np.nan

# feature 변환을 한 이후의 design matrix
Z = np.ones((N,1))
for D in range(1, 20):
    Z = np.hstack([Z, x**D]) # x0승부터 x19승까지
    
    theta = np.linalg.inv(Z.T.dot(Z)).dot(Z.T).dot(y)
    y_pred = Z.dot(theta)
    
    sse[D] = np.sum( (y - y_pred)**2 )    
    theta_norm2[D] = np.sum( theta**2 )
    
#     plt.figure()
#     plt.scatter(x,y,s = 10, c = 'gray')
#     plt.scatter(x,y_pred,s = 10, c = 'red')
    
plt.figure()
plt.subplot(2,1,1)
plt.plot(sse) #error가 줄어든다
plt.subplot(2,1,2)
plt.plot(theta_norm2) # 모델의 복잡도가 증가한다
```

```python
# regulalization 추가
N,D = x.shape
alpha = 1e1

sse = np.zeros(20)
sse[:] = np.nan

theta_norm2 = np.zeros(20)
theta_norm2[:] = np.nan

# ridge regression, normal equation
# feature 변환을 한 이후의 design matrix
Z = np.ones((N,1))
for D in range(1, 20):
    Z = np.hstack([Z, x**D])
    
    theta = np.linalg.inv(Z.T.dot(Z) + (alpha/2)*np.identity(D + 1)).dot(Z.T).dot(y)
    y_pred = Z.dot(theta)
    
    sse[D] = np.sum( (y - y_pred)**2 )    
    theta_norm2[D] = np.sum( theta**2 )
    
    plt.figure()
    plt.scatter(x,y,s = 10, c = 'gray')
    plt.scatter(x,y_pred,s = 10, c = 'red')
    
plt.figure()
plt.subplot(2,1,1)
plt.plot(sse)
plt.subplot(2,1,2)
plt.plot(theta_norm2)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso

ridge_alpha = 1e1
lasso_alpha = 1e0

linear = LinearRegression()
ridge = Ridge(alpha = ridge_alpha)
lasso = Lasso(alpha = lasso_alpha, fit_intercept = False)

Z = np.ones((N,1))
for D in range(1,10):
    Z = np.hstack([Z, x**D])

linear.fit(Z, y)
ridge.fit(Z, y)
lasso.fit(Z, y)

linear_norm2 = np.sum( linear.coef_**2 )
ridge_norm2 = np.sum( ridge.coef_**2 )
lasso_norm2 = np.sum( lasso.coef_**2 )

print('norm squre: linear %.2f ridge %.2f lasso %.2f' % (linear_norm2, ridge_norm2, lasso_norm2))

plt.figure()
plt.plot(linear.coef_[0])
plt.plot(ridge.coef_[0])
plt.plot(lasso.coef_)
plt.plot(np.array([5, 1, -2, 1e-1, 0, 1e-2]))
plt.legend(['linear','ridge','lasso','true'])
```
