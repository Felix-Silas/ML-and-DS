```python
import numpy as np
import matplotlib.pyplot as plt

# 코드 실행 시간 측정
# anaconda prompt에 pip install pytictoc해서 설치
from pytictoc import TicToc
```

```python
t = TicToc()
t.tic() #시간 측정 출발
# 여기 사이의 시간을 측정
t.toc() # 시간 측정 종료
```

```python
# 행렬을 쌓는 함수
# hstack = 가로로 horizontal
# vstack = 세로로 vertical

x = np.arange(10)
y = 2*x

print(x)
print(y)
print(x.shape)
print(y.shape)

# numpy 벡터 = 행 벡터가 디폴트

z1 = np.hstack([x, y])
z2 = np.vstack([x, y])

print(z1)
print(z2)

print(z1.shape) # 벡터 ndim = 1짜리인, shape = (20, )
print(z2.shape) # 1 x 10짜리 벡터 두개를, 세로로 쌓은듯한, 2 x 10짜리 행렬
```

```python
X = np.random.rand(10, 2) # uniform distribution, [0,1] 사이의 실수
print(X)

print(np.sum(X)) # 전체 값을 다 더함, 스칼라 한개
print(np.sum(X, axis = 0)) # 더하는 방향을 0번째 축, 행에 대해서 덧셈을 해달라, 결과는 열
print(np.sum(X, axis = 1)) # 더하는 방향을 1번째 축, 열에 대해서 덧셈을 해달라, 결과는 행

print(np.mean(X)) # 전체 값에 대한 평균
print(np.mean(X, axis = 0)) # 행에 대해서 평균, 위에서 아래
print(np.mean(X, axis = 1)) # 열에 대해서 평균, 오른쪽에서 왼쪽
```

```python
x = np.random.rand(10)

print(x)
print(np.max(x)) # 들어있는 값들 중에 제일 큰 값을 리턴
print(np.argmax(x)) # 들어있는 값들 중에 제일 큰값이 있는 인덱스를 리턴
print(np.min(x))
print(np.argmin(x))

# np.max(x) == x[np.argmax(x)]
# np.min(x) == x[np.argmin(x)]
print(x[np.argmax(x)])
print(x[np.argmin(x)])

print(np.where(x >= 0.5)) # 인덱스를 리턴
print(np.where(x >= 1))
```

```python
# 100개의 점을 만들고 1개의 빨간 x 기준점을 만드는 작업
N = 100 #샘플의 갯수
D = 2  # 차원

X = np.random.randn(N, D) # N x D, N개의 점, D차원의 벡터, 행벡터로 들어감
z = np.array([0.5,0.2]) #거리를 구할 기준점 1개

plt.scatter(X[:,0], X[:,1], s = 10)
plt.scatter(z[0], z[1], marker = 'x', facecolor = 'r')
```

```python
# 100개의 점과 빨간 점 사이의 거리를 모두 구하고, 가장 먼 점을 밝은 파란색으로 표시하기 (2차원일 때 한정, 안 그러면 for문 2개)
N = 100 
D = 2  

X = np.random.randn(N, D) 
z = np.array([0.5,0.2])

# X 안에 들어있는 n번째 점의 위치 = (X[n,0], X[n,1])
# z라는 기준점 = (z[0], z[1])
# (X[n,0] - z[0])**2 + (X[n,1] - z[1])**2

Dist = np.zeros(N)
for n in range(N):
    Dist[n] = np.sqrt((X[n,0]-z[0])**2 + (X[n,1]-z[1])**2)   # 여러 차원에 대응하려면 이걸 바꾸어야 함

id_Long = np.argmax(Dist)

plt.scatter(X[id_Long,0], X[id_Long,1], edgecolor = 'cyan', facecolor = 'w')
plt.scatter(X[:,0], X[:,1], s = 10)
plt.scatter(z[0], z[1], marker = 'x', facecolor = 'r')
```

```python
N = 100 # 샘플의 갯수
D = 5

X = np.random.randn(N, D) # N x D, N개의 점, D차원의 벡터, 행벡터로 들어감
z = np.random.rand(D) # 거리를 구할 기준점 한개

Dist1 = np.zeros(N)
Dist2 = np.zeros(N)
Dist3 = np.zeros(N)

#1번째 방법: for문 2개
# X 안에 들어있는 n번째 점의 위치 = (X[n,0], X[n,1], X[n,2], ... , X[n,d])
# z라는 기준점 = (z[0, z[1], z[2], ... , z[d])

t.tic()
for n in range(N):
    for d in range(D):
        Dist1[n] += (X[n,d] - z[d])**2
    Dist1[n] = np.sqrt(Dist1[n])
t.toc()

# 2번째 방법: for문 1개
# X[n, :] = (X[n,0], X[n,1], X[n,2], ... , X[n,d])
# (z[0, z[1], z[2], ... , z[d])
t.tic()
for n in range(N):
    Dist2[n] = np.sqrt(np.sum((X[n,:] - z)**2))
t.toc()

# 3번째 방법: for문 없이, X에 모인 점들과, z에 명시한 기준점, 사이의 거리
# X = N x D    z = 1 x D    Z = N x D
t.tic()
Z = np.matlib.repmat(z.reshape(1,D), N, 1)  # reshape 해주는 이유: 스칼라를 벡터로 바꿔주기 위해
Dist3 = np.sum( (X - Z)**2, axis = 1 )  # ------------> 이 방향으로 더해야 Dist3가 된다!
t.toc()
# array는 될 수 있으면 array인채로 계산을 하면 훨씬 빠름

idLong = np.argmax(Dist1) #제일 멀리있는 점
    
plt.scatter(X[idLong,0], X[idLong,1], s = 50, edgecolor = 'cyan', facecolor = 'w')
plt.scatter(X[:,0], X[:,1], s = 10)
plt.scatter(z[0], z[1], marker = 'x', facecolor = 'r')
```
