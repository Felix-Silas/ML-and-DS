```python
import numpy as np
import numpy.matlib
import math
import matplotlib.pyplot as plt

N = 1000
D = 2

X = np.random.randn(D,N)
A = np.random.randn(D,D)
X = A.dot(X)

plt.figure()
ax = plt.subplot(1,1,1)
plt.scatter(X[0,:], X[1,:], c = 'gray', s = 10)
ax.set_aspect('equal')
```

```python
C = np.cov(X) # X = D x N, C = D x D
print(C.shape)

U, s, V = np.linalg.svd(C, full_matrices = True)
# C = U * diag(s) * V
# 우리가 아는 꼴로 만들기 위한 작업 2개
S = np.diag(s)
V = V.T

u1 = U[:,0]
u2 = U[:,1]
s1 = s[0]
s2 = s[1]

plt.figure()
ax = plt.subplot(1,1,1)
plt.scatter(X[0,:], X[1,:], c = 'gray', s = 10)
ax.set_aspect('equal')

plt.plot([0, s1*u1[0]], [0, s1*u1[1]])
plt.plot([0, s2*u2[0]], [0, s2*u2[1]])
```
