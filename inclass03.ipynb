```python
import numpy as np
# [*] = 실행중 {1} = 실행완료, 몇번째
```

```python
# 새로운 array를 생성
# 1차원 = vector, 2차원 = matrix

a = np.array([1, 2, 3])  # 1차원 벡터를 생성
A = np.array([[1, 2], [3, 4]]) # 2x2의 2차원 행렬을 생성
B = np.array([[1, 2, 3], [4, 5, 6]]) #2x3의 2차원 행렬을 생성

print(a)
print(A)
print(B)
```

```python
# 0, 1 또는 특정한 값으로 초기화

X = np.zeros((3, 4)) # 3x4의 0으로 가득찬 행렬
Y = np.ones((2,5)) # 2x5의 1로 가득찬 행렬
Z = np.full((3,2), 100) # 첫 번째 인자의 shape으로, 두번째 인자로 가득찬 행렬

print(X)
print(Y)
print(Z)
```

```python
# np.arange(시작하는 값, 마지막 값(디폴트 = 포함 x), 증가량(디폴트 = 1))

x = np.arange(0, 10)
y = np.arange(0, 10, 2)

print(x)
print(y)
```

```python
# np.linspace(시작하는 값, 마지막 값(디폴트 = 포함 o), 전체 개수) 
#전체 개수개로 쪼개라

x = np.linspace(0, 10, 5)
y = np.linspace(0, 10, 5, endpoint = False)
z = np.linspace(0, 10, 5, endpoint = True)  # true가 기본값


print(x)
print(y)
print(z)
```

```python
# random으로 샘플을 생성
X = np.random.rand(2, 3)   # [0, 1] 사이의 uniform distribution을 따르는 랜덤 샘플
Y = np.random.randn(2, 3)  # Normal distribution (평균 = 0, 분산 = 1)을 따르는 랜덤 샘플

print(X)
print(Y)
```

```python
# np.ndarray에 있는 속성

z = np.ones(3)  # 1차원 array = 벡터
Z = np.ones((2, 3)) # 2차원 array = 행렬

print(z.ndim, z.shape, z.size)
print(Z.ndim, Z.shape, Z.size)

# ndim = 1 벡터, 2 행렬
# shape = (몇 행, ) or (몇 행, 몇 열)
# size = 전체 스칼라의 개수 = 행x열

print(z)
print(Z)
```

```python
# reshape

a = np.arange(0, 12) # 1차원 벡터, 12x1
b = a.reshape(4, 3) # a를 4x3으로 reshape
c = b.reshape(6, 2)
#d = c.reshape(5, 2) size가 12가 아니라서 안 된다 오류 발생!

# reshape할 때 값이 채워지는 순서를 잘 기억할 것!
print(a)
print(b)
print(c)
```

```python
# array의 값에 접근하는 방법

a = np.arange(0, 12)   # 1차원 벡터, 12x1
a = a.reshape(3, 4)    # a를 4x3으로 reshape

print(a)

# [행, 열]
# 시작값: 끝값 = [시작값, 시작값+1, .... , 끝값-1]
# -1 = 마지막 위치, -2 = 마지막에서 두번째 위치


print(a[2,3]) # 0부터 세므로 3행 4열 값이다
print(a[0:2, 3]) # a[[0,1],3] = 1행~2행, 4열
print(a[0:3, 3]) # a[[0,1,2],3] = 1행~3행, 4열

print(a[-1, -1]) # -1 = 마지막 위치 [-1, -1] = 마지막 행, 마지막 열
print(a[-2, -2]) # -2 = 마지막에서 두 번째 [-2, -2] = 마지막에서 두번째 행과 열

print(a[1:-1, -2]) # {1:-1, -2} = [1: 2, -2] = [1, -2]
print(a[1, 1: -1])  # [1, 1:-1] = [1, [1, 2]]

print(a[1:,2:])  # a의 2행~마지막행, 3열~마지막열
```
